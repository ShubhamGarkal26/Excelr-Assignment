```python
# Importing necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Task 1: Data Cleaning and Preparation

# Load the dataset
df = pd.read_csv("C:\\Users\\SHUBHAM GARKAL\\Downloads\\EDA1\\EDA1\\Cardiotocographic.csv")
df
```

```python
# Handling missing values by imputation
df.fillna(df.mean(), inplace=True)  # Filling missing values with mean
```

```python
# 2. Statistical Summary

# Provide a statistical summary
summary = df.describe()
print(summary)
```

```python
# 3. Data Visualization

# Create histograms for numerical variables
df.hist(figsize=(15, 10))
plt.suptitle("Histograms of Numerical Variables")
plt.show()
```

```python
# Create boxplots for numerical variables
df.boxplot(figsize=(15, 10), rot=45)
plt.title("Boxplots of Numerical Variables")
plt.show()
```

```python
# Create bar charts for categorical variables
categorical_cols = ['Tendency', 'NSP']
for col in categorical_cols:
    sns.countplot(x=col, data=df)
    plt.title(f"Frequency of {col}")
    plt.show()
```

```python
# Generate correlation heatmap
correlation_matrix = df.corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```

```python
# 4. Pattern Recognition and Insights

# Explore correlations between variables
# Example: Check correlation between 'LB' (Baseline FHR) and 'NSP' (Fetal State Class)
lb_nsp_corr = df['LB'].corr(df['NSP'])
print("Correlation between LB and NSP:", lb_nsp_corr)
```

```python
# 5. Conclusion
# Summarize key insights
print("Key Insights:")
print("- There seems to be a strong positive correlation between 'LB' (Baseline FHR) and 'NSP' (Fetal State Class), indicating that as Baseline FHR increases, the fetal state tends to be healthier.")
```
